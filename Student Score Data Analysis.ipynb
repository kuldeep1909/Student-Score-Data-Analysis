# to check the gender distribution

```python
plt.figure(figsize=(5, 5))
ax = sns.countplot(data = df, x = 'Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# the above chart concluded that:
#the number of female is more then number of male
```

# Parent education impact on student score

```python
gb = df.groupby('ParentEduc').aggregate({'MathScore' : 'mean', 'ReadingScore' : 'mean', 'WritingScore' : 'mean' })
gb
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(gb, annot = True)
plt.title('Gender Distribution')
plt.show()
```

```python
# the above chart cleared that the parenet education impact their students score
```

```python
gb1 = df.groupby('ParentMaritalStatus').aggregate({'MathScore' : 'mean', 'ReadingScore' : 'mean', 'WritingScore' : 'mean' })
gb1
```

```python
plt.figure(figsize=(4, 4))
sns.heatmap(gb1, annot = True)
plt.title('ParentMaritalStatus v/s Student Score')
plt.show()
```

```python
# the above chart says that the marital status of parents is negligible on the student scores
```

# Distribution of EthnicGroup

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

l = ['group A', 'group B', 'group C', 'group D', 'group E']
mylist = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
print(mylist)
plt.pie(mylist, labels=l, autopct='%1.2f%%')
plt.title('Distribution of Ethinic Groups')
plt.show()
```
